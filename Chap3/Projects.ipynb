P-3.55 Perform an experimental analysis of the three algorithms prefix average1,
prefix average2, and prefix average3, from Section 3.3.3. Visualize their
running times as a function of the input size with a log-log chart.

```python
import matplotlib.pyplot as plt


def prefix_average1(S):
    n = len(S)
    A = [0] * n
    for i in range(n):
        total = 0
        for j in range(i + 1):
            total += S[j]
        A[i] = total / (i + 1)
    return A


def prefix_average2(S):
    n = len(S)
    A = [0] * n
    for i in range(n):
        A[i] = sum(S[0:i + 1]) / (i + 1)
    return A


def prefix_average3(S):
    n = len(S)
    A = [0] * n
    total = 0
    for i in range(n):
        total += S[i]
        A[i] = total / (i + 1)
    return A


def time_prefix_average(func, n):
    import time
    start = time.time()
    S = [0] * n
    func(S)
    end = time.time()
    return end - start


N = [10, 100, 1000, 10000, 100000, 1000000]
# times for prefix_average1
time1 = [time_prefix_average(prefix_average1, n) for n in N]
# times for prefix_average2
time2 = [time_prefix_average(prefix_average2, n) for n in N]
# times for prefix_average3
time3 = [time_prefix_average(prefix_average3, n) for n in N]

# plot the results
plt.loglog(N, time1, label="prefix_average1")
plt.loglog(N, time2, label="prefix_average2")
plt.loglog(N, time3, label="prefix_average3")
plt.xlabel("Input size (N)")
plt.ylabel("Running time (seconds)")
plt.legend()
plt.show()
```

P-3.57 Perform experimental analysis to test the hypothesis that Pythonâ€™s sorted
method runs in O(nlogn) time on average.

```python
import math
import matplotlib.pyplot as plt
import time
# 1. Create a set of data of varying sizes to test the hypothesis. For example, use an array of size 10, 20, 30, 40, etc.

# Create an array of sizes 10, 20, 30, 40, 50
arr = [90, 20, 10, 60, 50]

# 2. Create a timer to measure the amount of time it takes to sort each set of data.

# Import the time library

# Create a start time
start_time = time.time()

# 3. Perform the sorting operation on each set of data and record the time it takes to complete.

# Perform sorting operation on each array
for i in arr:
    sorted_arr = sorted(arr[:i])
    end_time = time.time()
    total_time = end_time - start_time
    print("The time taken to sort the array of size {} is {}.".format(i, total_time))

# 4. Calculate the average time it takes to sort each set of data.

# Calculate the average time
total_time_arr = []
for i in arr:
    sorted_arr = sorted(arr[:i])
    end_time = time.time()
    total_time = end_time - start_time
    total_time_arr.append(total_time)

avg_time = sum(total_time_arr) / len(total_time_arr)
print("The average time taken to sort the array is {}".format(avg_time))

# 5. Plot the average time versus the size of the data set to determine if the time taken is consistent with an O(nlogn) time complexity.

# Import the matplotlib library

# Plot the average time vs size of the array
plt.plot(arr, total_time_arr, 'b-')
plt.xlabel('Size of the array')
plt.ylabel('Average time taken to sort')
plt.title('Average time taken to sort vs Size of the array')
plt.show()

# 6. Compare the plotted results with the expected O(nlogn) time complexity line to test the hypothesis.

# Compare the plotted results with the expected O(nlogn) time complexity line
x = arr
y = [x * (math.log(x)) for x in arr]
plt.plot(x, y, 'r--')
plt.legend(['Actual Time Taken', 'O(nlogn) Time Complexity'])
plt.xlabel('Size of the array')
plt.ylabel('Average time taken to sort')
plt.title('Average time taken to sort vs Size of the array')
plt.show()

# From the comparison of the plotted results, we can conclude that the Pythonâ€™s sorted method runs in O(nlogn) time on average.
```
